# Lemmatization in NLP
Lemmatization is the process of converting a word to its base or dictionary form (called lemma) using vocabulary and grammar rules.
Unlike stemming, it does not just cut letters.
It understands the wordâ€™s meaning and part of speech.


```python
words = ["eating","eaten","eat","happy","happyness","running","ran","goen","going","go"]
words
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

words = ["eating","eaten","eat","happy","happyness","running","ran","goen","going","go"]

lemmatizer = WordNetLemmatizer()

for word in words:
    print(word + " ----> " + lemmatizer.lemmatize(word, pos='v'))
```
