```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')

words = word_tokenize(sentence)

print(words)
```

```python
import nltk
nltk.download('words')
tagg= nltk.pos_tag(sentence)

nltk.ne_chunk(tagg).draw()
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")

text = """
Artificial Intelligence is transforming industries across the world. 
Healthcare systems use AI for disease detection and personalized treatment. 
Financial institutions rely on machine learning to detect fraud and manage risk. 
Education platforms use AI to personalize learning experiences. 
However, AI systems also raise concerns about privacy, bias, and transparency.
"""

summary = summarizer(text, max_length=60, min_length=20, do_sample=False)

print(summary[0]['summary_text'])
```
